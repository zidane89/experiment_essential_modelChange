```python
import matplotlib.pyplot as plt 
import pickle 
```

```python
with open("DDQN3_1.pkl", "rb") as f: 
    data_w = pickle.load(f)[10]
with open("DDQN3_2.pkl", "rb") as f: 
    data_kw = pickle.load(f)[10]
with open("DDQN3_3.pkl", "rb") as f: 
    data_norm = pickle.load(f)[10]
    
```

# visualization W vs KW 

```python
plt.figure(figsize=(12, 7))
plt.plot(data_w["rewards"], label="power in W", linewidth=2) 
plt.plot(data_kw["rewards"], label="power in kW", linewidth=2)

plt.xlabel("episodes", fontsize=16)
plt.ylabel("episodic rewards", fontsize=16)
plt.xlim([0, 200])
plt.legend(fontsize=14, loc="upper left") 
plt.grid() 
# plt.ylim([-1000, -250])
plt.savefig("rewards_with_states.png")
```

# visualization W vs KW vs normalization 

```python
plt.figure(figsize=(12, 7))
plt.plot(data_w["rewards"], label="power in W", linewidth=2) 
plt.plot(data_kw["rewards"], label="power in kW", linewidth=2)
plt.plot(data_norm["rewards"], label="power in normalization", linewidth=2)

plt.xlabel("episodes", fontsize=16)
plt.ylabel("episodic rewards", fontsize=16)
plt.xlim([0, 200])
plt.legend(fontsize=14, loc="upper left") 
plt.grid() 
# plt.ylim([-1000, -250])
plt.savefig("rewards_with_states.png")
```
